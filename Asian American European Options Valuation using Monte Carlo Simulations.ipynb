## Simulating Geometric Brownian Motion


The Black-Scholes-Merton model is given by the SDE:

$$
dS_t = rS_tdt + \sigma S_t dZ_t
$$

An Euler discretization of the model for MCS :

$$
S_t = S_{t-\Delta t} \cdot \exp \left(\left(r - \frac{\sigma^2}{2}\right)\Delta t + \sigma \sqrt{\Delta t} z_t\right)
$$



```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
S0 = 100  # initial value
r = 0.05  # constant short rate
sigma = 0.25  # constant volatility
T = 2.0  # in years
I = 10000  # number of random draws
```

```python
I = 10000
M = 50
dt = T / M
S = np.zeros((M + 1, I))
S[0] = S0


for t in range(1, M + 1):
    S[t] = S[t - 1] * np.exp((r - 0.5 * sigma ** 2) * dt 
            + sigma * np.sqrt(dt) * np.random.standard_normal(I))
```

## European Option Valuation using the Mone Carlo Simulations

*** Valuation of European call option in Black-Scholes-Merton by Monte Carlo simulation (of index level at maturity) ***
    

```python
S0 = 100. #Stock Price
r = 0.05 #risk free rate
sigma = 0.25 #Stock Constant Volatility
T = 1.0 #Time for expiry
I = 50000 #Number of simulations
M = 50 #Time Steps for simulations
dt = T/M #discrete time interval
```

*** "european_option_pricing_using_gbm" function takes Strike Price and option type as parameter and returns expected value of pay off ***

```python
def european_option_pricing_using_gbm(K, option='call'):

    S = np.zeros((M + 1, I))    
    S[0] = S0
    
    for t in range(1, M + 1):
        S[t] = S[t - 1] * np.exp((r - 0.5 * sigma ** 2) * dt 
                + sigma * np.sqrt(dt) * np.random.standard_normal(I))
    
    # calculate payoff at maturity
    if option == 'call':
        payOff = np.maximum(S[-1] - K, 0)
    else:
        payOff = np.maximum(K - S[-1], 0)
    
    # calculate MCS estimator
    optionPrice = np.exp(-r * T) * 1 / I * np.sum(payOff )
    
    return optionPrice
```

```python
print("European Option Price for Strike 100 : {}".format(european_option_pricing_using_gbm(100)))
```

```python
eurCallPayOffList = []
eurPutPayOffList = []

for K in np.arange(50, 151, 10):
    eurCallPayOffList.append(european_option_pricing_using_gbm(K,'call'))
    eurPutPayOffList.append(european_option_pricing_using_gbm(K,'put'))
```

```python
plt.title('European Call Option')
plt.plot(np.arange(50, 151, 10), eurCallPayOffList)
plt.plot(np.arange(50, 151, 10),eurCallPayOffList, 'ro')
plt.xlabel('Strike Price')
plt.ylabel('Option Value')
```

## American Option Valuation using the Least Square Monte Carlo ( LSM ) Algorithm

Download the paper by Longstaff and Schwartz (2001) about the Least-Squares Monte Carlo algorithm to price American options from [Paper about LSM algorithm](https://people.math.ethz.ch/~hjfurrer/teaching/LongstaffSchwartzAmericanOptionsLeastSquareMonteCarlo.pdf)

```python
def american_option_pricing_using_gbm(K, option='call'):

    S = np.zeros((M + 1, I))    
    S[0] = S0
    df = np.exp(-r*dt)
    
    for t in range(1, M + 1):
        S[t] = S[t - 1] * np.exp((r - 0.5 * sigma ** 2) * dt 
                + sigma * np.sqrt(dt) * np.random.standard_normal(I))
    
    # calculate payoff at maturity
    if option == 'call':
        payOff = np.maximum(S - K, 0)
    else:
        payOff = np.maximum(K - S, 0)
        
    # LSM algorithm
    V = np.copy(payOff)
    for t in range(M - 1, 0, -1):
        reg = np.polyfit(S[t], V[t + 1] * df, 7) # Using 7 degree of freedom to find the continuation value for Option Price
        C = np.polyval(reg, S[t])
        
        # If Continuation value > Payoff then take discounted value of pay off at T+1 else take pay off value at time T
        V[t] = np.where(C > payOff[t], V[t + 1] * df, payOff[t]) 
    
    # calculate MCS estimator
    optionPrice = df * 1 / I * np.sum(V[1])
    
    return optionPrice
```

```python
american_option_pricing_using_gbm(100)
```

```python
usCallPayOffList = []
usPutPayOffList = []

for K in np.arange(50, 151, 10):
    
    usCallPayOffList.append(american_option_pricing_using_gbm(K,'call'))
    usPutPayOffList.append(american_option_pricing_using_gbm(K,'put'))
```

```python
plt.title('European vs American Call Option')
plt.plot(np.arange(50, 151, 10), eurCallPayOffList)
plt.plot(np.arange(50, 151, 10),usCallPayOffList, 'ro')
plt.xlabel('Strike Price')
plt.ylabel('Option Value')
```

```python
plt.title('European vs American Put Option')
plt.plot(np.arange(50, 151, 10), eurPutPayOffList)
plt.plot(np.arange(50, 151, 10),usPutPayOffList, 'ro')
plt.xlabel('Strike Price')
plt.ylabel('Option Value')
```

*** Difference between european and American option is early exercise premium ***

### Asian Options using Monte Carlo Simulations

Asian options uses average price of the Stock path for full time till maturity therefore they are cheaper than european options 

*** Lets see this in practical ***

```python
def asian_option_pricing_using_gbm(K, option='call'):

    S = np.zeros((M + 1, I))    
    S[0] = S0
    
    for t in range(1, M + 1):
        S[t] = S[t - 1] * np.exp((r - 0.5 * sigma ** 2) * dt 
                + sigma * np.sqrt(dt) * np.random.standard_normal(I))
    
    # calculate payoff at maturity
    if option == 'call':
        payOff = np.maximum(S.mean(axis=0) - K, 0) # Average of Stock Price taken on path for each dt timestamp
    else:
        payOff = np.maximum(K - S.mean(axis=0), 0) # Average of Stock Price taken on path for each dt timestamp
    
    # calculate MCS estimator
    optionPrice = np.exp(-r * T) * 1 / I * np.sum(payOff )
    
    return optionPrice
```
